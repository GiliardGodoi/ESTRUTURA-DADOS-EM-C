```python
import sys
import os

diretorio_imagens = os.path.join('..','img')
diretorio_dados = os.path.join('..','data')
arquivo = os.path.join(diretorio_dados,'opt_griewank_function_gbest.csv')

## Alterar o valor desta variavel para salvar imagens com o nome da simulacao apropriado
nome_simulacao = 'griewank'
```

```python
os.listdir(diretorio_dados)
```

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_dados = pd.read_csv(arquivo,index_col=0,header=0)
```

```python
df_dados['fitness'].head(10)
```

```python
ga_population_size = 10
ga_repetitions = 50

df_fitness = df_dados['fitness']

best_fitness = [ min(df_fitness.iloc[i:(i+ga_population_size)]) for i in range(0,len(df_dados),ga_population_size) ]

print(len(best_fitness))

best_fitness[0]
```

```python
ga_population_size = 10
ga_repetitions = 50

df_fitness = df_dados['fitness']

mean_fitness = [ np.mean(df_fitness.iloc[i:(i+ga_population_size)]) for i in range(0,len(df_dados),ga_population_size) ]

print(len(mean_fitness))

mean_fitness[0]
```

```python
plt.figure()
plt.plot(best_fitness)

plt.title('Melhor fitness para cada geração')
plt.xlabel('GA gerações')
plt.ylabel('Best Fitness')
# plt.yscale('symlog', linthreshy=0.01)

plt.grid(True)

salvar_imagem_em = os.path.join(diretorio_imagens,f'best_fitness_{nome_simulacao}.png')
plt.savefig(salvar_imagem_em)
```

```python
plt.figure()
plt.plot(mean_fitness)

plt.title('Média fitness para cada geração')
plt.xlabel('GA gerações')
plt.ylabel('Fitness Médio')
# plt.yscale('symlog', linthreshy=0.01)

plt.grid(True)

salvar_imagem_em = os.path.join(diretorio_imagens,f'fitness_medio_{nome_simulacao}.png')
plt.savefig(salvar_imagem_em)
```
